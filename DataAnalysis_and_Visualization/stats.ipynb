```python
#using numpy
median = np.median(list_)
print(median)

#using math
def median_fun(lis):
    sorted_list = sorted(lis)
    n_ = len(list_)
    if n_%2!=0:
        median = sorted_list[int((n_+1)//2)-1]
        return median
    else:
        median = (sorted_list[int((n_//2)-1)]+sorted_list[int(((n_//2)+1)-1)])/2
        return median
median_ = median_fun(list_)
print(median_)
```

## Variance


In probability theory and statistics, variance is the expectation of the squared deviation of a random variable from its mean. In other words, it measures **how far a set of numbers is spread out from their average value**.

When you have "N" data values that are:

1. The Population: divide by N when calculating Variance
2. A Sample: divide by N-1 when calculating Variance

```python
#using numpy
variance = np.var(list_)
print(variance)

#using math
temp = []
for i in range(len(list_)):
    temp.append((list_[i]-mean)**2)
variance_ = sum(temp)/(n_-1)
print(variance)
```

## Function for calculating percentile

```python
#function for calculating percentile
def percentile(lis,p):
    #1 Sort the list
    sort_list = sorted(lis)
    n = len(sort_list)
    #2 calculating location of p_th percentile:lp
    lp = (p/100)*(n-1)
    #3 calculating the integer part:ip and fractional part:fp
    ip = math.floor(lp)
    fp=lp-ip
    #4 computing p_th percentile:percentile
    percentile = sort_list[ip] + ( fp * (sort_list[ip+1]-sort_list[ip]))
    return percentile
```

## IQR (Inter-Quartile Range)



```python
#using scipy
print(stats.iqr(list_))

#using math 
# IQR = Q3-Q1
Q1 = percentile(list_,25)
Q3 = percentile(list_,75)
IQR = Q3-Q1
print(IQR)
    
    
```
